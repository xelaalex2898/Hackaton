```python
import requests

url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=19.351423%2C-99.069973&radius=300&type=restaurant&key=token"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
```

```python
y=json.loads(response.text)

names = []
latituds= []
longituds = []

for i in range(len(y['results'])):
    names.append(y['results'][i]["name"])
    latituds.append(float(y['results'][i]['geometry']['location']['lat']))
    longituds.append(float(y['results'][i]['geometry']['location']['lng']))
    print(y['results'][i]["name"],"-",y['results'][0]['geometry']['location']['lat'],y['results'][0]['geometry']['location']['lng'])
```
